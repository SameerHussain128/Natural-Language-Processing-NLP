```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image
```

```python
text2 = '''Python is a versatile and powerful programming language known for its simplicity and readability. With Python, you can automate tasks, build web applications, analyze data, and much more. It has a large and active community that contributes to its extensive library ecosystem, making it easy to find and utilize various functionalities.
In Python, you can work with strings, lists, dictionaries, and other data structures effortlessly. The syntax is clean and intuitive, allowing you to write code that is both concise and expressive. Additionally, Python supports object-oriented programming, functional programming, and other programming paradigms, giving you flexibility in how you approach problem-solving.'''
```

```python
mask = np.array(PIL.Image.open(r'C:\Users\SAMEER\OneDrive\Desktop\Data Science 6pm\pc.png'))

colormap = ImageColorGenerator(mask)

wc = WordCloud(stopwords=set(list(STOPWORDS)+['support']),
               mask = mask,
               background_color= 'white', contour_color = 'black',contour_width=3,
               min_font_size=3,max_words=400).generate(text2)
wc.recolor(color_func = colormap)
plt.imshow(wc)
plt.axis("off")
plt.show()
```
